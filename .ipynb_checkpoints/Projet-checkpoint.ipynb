```python
# Importer les bibliotheqes necessaires 
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# recuperation de l'ensemble de données creditcard.csv
data = pd.read_csv('creditcard.csv',header=0)
```

```python
# visualisation des 5 premieres lignes du dataset
data.head()
```

```python
# l'ensemble de données a 31 caracteristique ou (colonnes)
# affichage de la taille de l'ensemble 
data.shape
```

```python
# l'ensemble de données contient 31 colonnes et 284807 enregistrement (ou lignes)
# data.info() pour  afficher le résumé des informations de l'ensemble de données, 
# telles que le nombre de lignes et de colonnes, 
#le type de données de chaque colonne, le nombre de valeurs non nulles, etc. 
data.info()
```

```python
# describe() affiche les informations statistique  de chaque colonne numerique de l'ensemble de données. 
#cette fonction permet de calculer les statistiques de base telles que la moyenne, l'écart type, la valeur minimale, la valeur maximale etc, 
#et le nombre de valeurs non nulles pour chaque colonne. 
data.describe()
```

```python
# trouver le nombre de valeurs manquantes ou nulles pour chaque colonne dans notre ensemble de données.

data.isnull().sum()
```

```python
#affichage de la matrice de correlation
# la matrice de correlation mesure la relation linéaire entre les variables de l'ensemble 

plt.figure(figsize=(30,20))
sns.heatmap(data.corr(),annot=True,cmap='YlOrRd')
plt.show()
```

```python
# dans notre cas nous allons utiliser toutes les variable de l'ensemble 
# donc on va pas selectionner les variables les plus correlées
# maintenant on va voir la distribution des valeurs des differente caracteristique de l'ensemble de données 
data.hist(figsize=(20,20))
plt.show()
```

```python
# la distribution des classes dans l'ensemble de données 
# on va commencer par le nombre d'observation de chaque classes 
data['Class'].value_counts()
```

```python
# nous remarquons que nous avons 2 classes 0 signifie qu'il y a pas de fraude 
# 1 signifie qu'il s'agit d'une fraude
# nous allons chimatiser cette distribution de classe 
plt.figure(figsize=(8,8))
plt.pie(data.Class.value_counts(),labels=data.Class.unique(),autopct='%0.2f%%')
plt.title('Distribution des classes')
plt.legend(loc='best')

plt.show()
```

```python
# nous remarquons que la classe 0 est la classe majoritaire avec 99,83% et 0,17% pour la classe 1 donc la classe minoritaire 
# on appelle ce genre d'ensemble de données un ensemble de données desiquilibré 
# a cause de la distribution non egale entre les enregistrement des differentes classes de l'ensemble 
# pour resoudre ce probleme il existe beaucoup de methodes re-échantillonage de l'ensemble de données 
# par exemple oversampling, undersimpling , SMOTE, ADASYN 
# dans ce dataset les echantillons de la classe minoritaire sont rare (492 echantillons)en comparaisant avec la classe majoritaire (284315)
# ce qui rend ces methodes presque inefficace 
# dans ce projet nous allons reduire le nombre d'echantillons dans la classe majoritaire ou 0 
# en premier nous allons copier notre dataframe data dans un autre df1
df1 = data.copy()
# on recupere les echantillons qui representent la classe majoritaire
df1 = df1.query('Class == 0')
```

```python
# on verifie les classe existante dans le nouveau dataframe
df1['Class'].value_counts()
```

```python
# on remarque qu'il ya que la classe 0 avec 284315 observation 
#nous allons afficher les 5 premieres ligne du nouveau dataframe
df1.head()
```

```python
# ensuite on prend 10000 observations au lieu de 284315
# comme ça on a pu reduire le nombre d'observation de la classe majoritaire 
df1 = df1.sample(n=10000)
```

```python
# on verifie la taille du dataframe 
df1.shape
```

```python
# par la suite on recupere les observation de la classe minoritaire 
df2 = data.copy()
df2 = df2.query('Class == 1')
```

```python
df2.head()
```

```python
# verifions les classe de ce dataframe 
df2['Class'].value_counts()
```

```python
# il contient une seule classe avec 492 observations
df2.shape
```

```python
# maintenant nous allons regrouper les deux nouveaux dataframe
df = pd.concat([df1,df2])
```

```python
# la taille de l'ensemble est la somme des observations des deux dataframe df1 et df2
df.shape
```

```python
df.head()
```

```python
# on sépare les entrées et les sorties de l'ensemble 
X = df.drop(columns=["Class","Time"])
y = df["Class"]
```

```python
from sklearn import preprocessing
scaler1 = preprocessing.StandardScaler().fit(X)
X=scaler1.transform(X) 
```

```python
# on divise l'ensemble de données en ensemble d'entrainement et ensemble de test
# l'ensemble d'entrainement (train) 
# l'ensemble de test pour evaluer le modele

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# modelisation avec GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
cassif = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0)
cassif.fit(X_train, y_train)
```

```python
# cette partie permet de trouver les prediction du modele en train et en test 
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix,accuracy_score
y_train_pred_b = cassif.predict(X_train)
y_test_pred_b = cassif.predict(X_test)
```

```python
# calcul de la precision globale ou bien accuracy du modele 
boost_train = accuracy_score(y_train, y_train_pred_b)
boost_test = accuracy_score(y_test, y_test_pred_b)
print(boost_train,boost_test)
```

```python
# calcul la precision du modele 
precision_train = precision_score(y_train, y_train_pred_b, average='macro')
precision_test = precision_score(y_test, y_test_pred_b, average='macro')
print(precision_train,precision_test)
```

```python

# on reprend notre modele avec les meilleurs parametre 
from sklearn.ensemble import GradientBoostingClassifier
cassif = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1,max_depth=1, random_state=0)
cassif.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix,accuracy_score
y_train_pred_m = cassif.predict(X_train)
y_test_pred_m = cassif.predict(X_test)
```

```python
# accuracy
accuracy_train = accuracy_score(y_train, y_train_pred_m)
accuracy_test = accuracy_score(y_test, y_test_pred_m)
print(accuracy_train,accuracy_test)
```

```python
# precision
precision_train = precision_score(y_train, y_train_pred_m, average='macro')

precision_test = precision_score(y_test, y_test_pred_m, average='macro')
print(precision_train,precision_test)
```

```python
print (precision_train)
```

```python
import matplotlib.pyplot as plt


# Définir les étiquettes pour les barres du graphique
bar_labels = ['Train', 'Test']

# Définir les valeurs pour les barres du graphique
bar_values_accuracy = [accuracy_train, accuracy_test]
bar_values_precision = [precision_train, precision_test]

# Créer le premier graphique en barres pour l'exactitude
plt.figure()
plt.bar(bar_labels, bar_values_accuracy, width=0.1,color='r')
plt.title("Accuracy du modéle")
plt.xlabel("ensemble")
plt.ylabel("Accuracy")

# Créer le deuxième graphique en barres pour la précision
plt.figure()
plt.bar(bar_labels, bar_values_precision, width=0.1, color='g')
plt.title("Precision du modéle")
plt.xlabel("ensemble")
plt.ylabel("Precision")

# Afficher les deux graphiques
plt.show()
```

```python
# le rapport de classification 
# le resumé de la classification du modele 
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred_m))
```

```python
# matrice de confusion 

cm = confusion_matrix(y_test, y_test_pred_m)
np.set_printoptions(precision=4)
sns.set(rc={'figure.figsize':(12,6)})
sns.heatmap(cm,annot=True, fmt="d",cmap='YlOrRd')
```

```python

# validation croisée

from sklearn.model_selection import cross_val_score
from sklearn import metrics
accuracy = cross_val_score(cassif, X_test, y_test, cv=3, scoring='accuracy')
print("Accuracy: %0.5f (+/- %0.5f)" % (accuracy.mean(), accuracy.std() * 2))
print("cv_scores: ", accuracy)
print('cv_scores mean:{}'.format(np.mean(accuracy)))
```

```python
precision = cross_val_score(cassif, X_test, y_test, cv=3, scoring='precision_macro')
print("Precision: %0.5f (+/- %0.5f)" % (precision.mean(), precision.std() * 2))
#print each cv score (accuracy) and average them
print("cv_scores: ", precision)
print('cv_scores mean:{}'.format(np.mean(precision)))
```

```python
recall = cross_val_score(cassif,X_test, y_test, cv=5, scoring='recall')
print("Recall: %0.5f (+/- %0.5f)" % (recall.mean(), recall.std() * 2))
print("cv_scores: ", recall)
print('cv_scores mean:{}'.format(np.mean(recall)))
```

```python
f = cross_val_score(cassif, X_test, y_test, cv=5, scoring='f1')
print("F-measure: %0.5f (+/- %0.5f)" % (f.mean(), f.std() * 2))
print("cv_scores: ", f)
print('cv_scores mean:{}'.format(np.mean(f)))
```

```python


import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

fpr, tpr, threshold = roc_curve(
    y_true=y_test, y_score=y_test_pred_m
)
plt.plot(fpr, tpr)
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.show()
```
